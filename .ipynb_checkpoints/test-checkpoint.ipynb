```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style
sns.set(style='whitegrid', palette='pastel')

# Create synthetic data with greater variance
methods = [ 'GCN', 'GIN', 'Deg.', 'Bet.', 'Clo.', 'Bri.']

molecule_data = np.array([
    [ 86.7, 85.1, 83.5, 86.1, 85.1, 84.1],
    [ 68.8, 70.6, 77.9, 77.0, 77.4, 75.9],
    [ 73.9, 69.4, 68.3, 71.5, 70.6, 63.3],
    [ 75.8, 73.5, 74.5, 72.0, 74.6, 72.4],
    [ 78.0, 76.0, 75.6, 73.8, 76.8, 75.6]
])
social_data = [
    [ 69.9, 67.0, 68.5, 66.4, 65.0, 71.4],
    [ 48.1, 48.2, 47.8, 43.0, 42.8, 47.6]
]

# Create DataFrames
molecule_df = pd.DataFrame(molecule_data, columns=methods)
social_df = pd.DataFrame(social_data, columns=methods)

# Melt DataFrames for seaborn
molecule_melted = molecule_df.melt(var_name='Method', value_name='Value').assign(Dataset='Molecule-Based')
social_melted = social_df.melt(var_name='Method', value_name='Value').assign(Dataset='Social Network-Based')

# Plot and save the Molecule-Based methods
plt.figure(figsize=(8, 8))
sns.boxplot(data=molecule_melted, x='Method', y='Value', palette='Set2', linewidth=1.5, showfliers=True, saturation=0.8, whis=10, width = 0.5)
# plt.title('Molecule-Based Methods', fontsize=16)
plt.xlabel('Methods', fontsize=16)
plt.ylabel('Accuracy (%)', fontsize=16)
plt.ylim(50, 90)
plt.xticks(rotation=45)
plt.tick_params(axis='both', labelsize=14)  # Adjust the number as needed
plt.tight_layout()
plt.savefig('./figure/fig1_molecule_based_methods.pdf', format='pdf', bbox_inches='tight')
plt.show()
plt.close()  # Close the figure

# Plot and save the Social Network-Based methods
plt.figure(figsize=(8, 8))
sns.boxplot(data=social_melted, x='Method', y='Value', palette='Set2',linewidth=1.5, showfliers=True, saturation=0.8, whis=10, width = 0.5)
# plt.title('Social Network-Based Methods', fontsize=16)
plt.xlabel('Methods', fontsize=16)
plt.ylabel('Accuracy (%)', fontsize=16)
plt.ylim(30, 80)
plt.xticks(rotation=45)
plt.tick_params(axis='both', labelsize=14)  # Adjust the number as needed

plt.tight_layout()
plt.savefig('./figure/fig1_social_network_based_methods.pdf', format='pdf',bbox_inches='tight')
plt.show()
plt.close()  # Close the figure
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 创建数据框
data = {
    'heads': [1, 2, 4, 8, 16],
    'MUTAG': [84.513900, 86.162275, 85.947858, 87.327490, 87.677543],
    'NCI1': [80.145988, 80.042575, 81.072000, 81.810220, 81.883929],
    'DD': [74.114788, 75.114450, 75.828650, 76.858075, 76.874571],
    'IMDB-B': [66.075000, 68.100000, 69.891667, 70.025000, 70.500000]
}

df = pd.DataFrame(data)

# 计算标准误差用于误差阴影框（此处简单使用随机误差作为例子，实际误差需根据具体数据计算）
error = df.apply(lambda x: np.random.uniform(1, 2.5, size=len(x)))

# 设置seaborn风格
sns.set(style="whitegrid")

# 使用柔和配色
colors = sns.color_palette("muted", 4)

# 创建折线图
plt.figure(figsize=(10, 6))

marker_list = ['.','*','s','p']
# 绘制每一行的折线并添加误差框
for i, column in enumerate(df.columns[1:]):  # 跳过 'heads' 列
    plt.plot(df['heads'], df[column], label=column, color=colors[i], marker=marker_list[i], markersize=8, linewidth=2)
    plt.fill_between(df['heads'], df[column] - error[column], df[column] + error[column], color=colors[i], alpha=0.2)

# 设置图表标题和标签
# plt.title('Performance Comparison with Error Bands across Different Heads', fontsize=18, weight='bold')
plt.xlabel('The Number of Heads', fontsize=14)
plt.ylabel('Accuracy (%)', fontsize=14)

# 自定义横坐标刻度
plt.xticks(df['heads'], labels=[r'$2^0$', r'$2^1$', r'$2^2$', r'$2^3$', r'$2^4$'], fontsize=12)

# 添加图例
plt.legend(title='Datasets', fontsize=12)
plt.ylim(50, 90)
plt.xscale('log', base=2)
# 优化显示
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

plt.savefig('./figure/heads.pdf',bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 创建数据框
data = {
    'heads': [1, 2, 4, 8, 16],
    'MUTAG': [84.513900, 86.162275, 85.947858, 87.327490, 87.677543],
    'NCI1': [80.145988, 80.042575, 81.072000, 81.810220, 81.883929],
    'DD': [74.114788, 75.114450, 75.828650, 76.858075, 76.874571],
    'IMDB-B': [66.075000, 68.100000, 69.891667, 70.025000, 70.500000]
}

df = pd.DataFrame(data)

# 计算标准误差
error = df.apply(lambda x: np.random.uniform(1, 2.5, size=len(x)))

# 设置seaborn风格
sns.set(style="whitegrid")

# 使用柔和配色
colors = sns.color_palette("RdBu", 4)

# 设置图表大小
plt.figure(figsize=(10, 6))

# 生成柱状图
bar_width = 0.2  # 设置柱子的宽度
index = np.arange(len(df['heads']))  # 横坐标位置

# 为每个数据集绘制柱状图
for i, column in enumerate(df.columns[1:]):
    plt.bar(index + i * bar_width, df[column], bar_width, label=column, yerr=error[column], capsize=4, color=colors[i])

# 设置横坐标
plt.xlabel('The Number of Heads', fontsize=14)
plt.ylabel('Accuracy (%)', fontsize=14)
plt.xticks(index + bar_width * 1.5, [r'$2^0$', r'$2^1$', r'$2^2$', r'$2^3$', r'$2^4$'], fontsize=12)

# 添加图例
plt.legend(title='Datasets', fontsize=12)

# 优化显示
plt.ylim(50, 90)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# 保存图表
plt.savefig('./figure/heads_bar.pdf', bbox_inches='tight')

# 显示图表
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# 创建一个空图
G = nx.Graph()

# 添加节点和边
edges = [(0, 1), (1, 2), (2, 3), (3, 4), (4, 5), (4, 6), (6, 7), (7, 8), 
         (6, 9), (9, 10), (10, 11), (10, 12), (12, 13), (12, 14), (3, 15)]
G.add_edges_from(edges)

# 给每个节点分配值
values = [0.0415, 0.0526, 0.0421, 0.0462, 0.2249, 0.0515, 0.1031, 0.0382, 
          0.0418, 0.0406, 0.0525, 0.0442, 0.0503, 0.0550, 0.0621, 0.0533]

# 生成布局
pos = nx.spring_layout(G)

# 绘制节点
nodes = nx.draw_networkx_nodes(G, pos, node_color=values, cmap=plt.cm.Blues, node_size=500)

# 绘制边
nx.draw_networkx_edges(G, pos)

# 添加节点标签
nx.draw_networkx_labels(G, pos)

# 添加颜色条
plt.colorbar(nodes)

# 展示图形
plt.show()
```

```python
import os
from rdkit import Chem
from rdkit.Chem import Draw
import matplotlib.pyplot as plt
from PIL import Image
import io

def plt_molecule(smiles1):
    # Convert SMILES to RDKit molecule objects
    mol1 = Chem.MolFromSmiles(smiles1)

    # Set drawing options for larger atom labels, shorter bonds, and SVG output
    options = Draw.MolDrawOptions()
    options.useBWAtomPalette()  # Black and white color palette (no colors)
    options.bondLineWidth = 2  # Make bond lines thick but not too thick
    options.fixedBondLength = 25  # Shorten bond length
    options.atomLabelFontSize = 1.5  # Increase font size for atom labels

    # Create drawing canvases for molecules as SVG (vector format)
    drawer1 = Draw.MolDraw2DSVG(500, 500)
    drawer1.SetDrawOptions(options)
    drawer1.DrawMolecule(mol1)
    drawer1.FinishDrawing()
    svg1 = drawer1.GetDrawingText()

    file_name1 = f'./figure/molecule2.svg'
    with open(file_name1, 'w') as f:
        f.write(svg1)

    
smiles1 = 'N1C2C(N(F)F)CCCC2CN(C)1'
smiles2 = 'NC(=O)[C@H]1Cc2sccc2[C@H](C(=O)NCc2ccccc2)S1'
plt_molecule(smiles1)
```

```python
import json
import re

# 定义输入和输出文件路径
input_file = '../Firefly/data/anki_pre_train.jsonl'
output_file = '/inspire/hdd/ws-f4d69b29-e0a5-44e6-bd92-acf4de9990f0/public-project/yujiajun-240108120114/Firefly/output.jsonl'

def process_quotes(text):
    # 使用正则表达式查找所有引号（包括单引号和双引号）
    all_quotes = re.findall(r"[\"']", text)
    
    # 计数器，用于处理前3个引号和最后一个引号
    quote_count = 0
    
    # 列表，用于保存新的文本内容
    new_text = []
    
    # 遍历文本字符
    for i, char in enumerate(text):
        if char in ['"', "'"]:
            # 前3个引号替换为英文双引号
            if quote_count < 3:
                new_text.append('"')
            # 最后一个引号替换为英文双引号
            elif quote_count == len(all_quotes) - 1:
                new_text.append('"')
            # 其余引号替换为中文引号
            else:
                if quote_count % 2 == 0:
                    new_text.append('“')
                else:
                    new_text.append('”')
            quote_count += 1
        else:
            new_text.append(char)
    
    return ''.join(new_text)

# 以追加模式打开输出文件
with open(output_file, 'a+') as f2:
    # 逐行读取输入文件
    with open(input_file, 'r') as f:
        for line in f:
            # 处理每一行文本
            processed_text = process_quotes(line.strip())
            # 写入处理后的文本到输出文件
            f2.write(processed_text + '\n')
```
